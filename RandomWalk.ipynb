```python
from random import uniform
import matplotlib.pyplot as plt
```

```python
step_lists = []
for j in range(5):
    value = 0
    step_lists.append([0])
    for i in range(1, 100):
        step = uniform(0, 1)
        if step > 0.5:
            value += 1
        else:
            value -= 1
        step_lists[j].append(value)
```

```python
fig, ax = plt.subplots()
for i in range(5):
    ax.plot(list(range(0, 100)), step_lists[i])
ax.set_xlabel("Steps during the random walk")
ax.set_ylabel("Value of the step")
ax.set_title("Simulation of 5 random walks")
```
